## Problem1 Markov Chain and Ranking

```python
data = pd.read_csv("CFB2019_scores.csv", header = None)
data.columns = ['Team_A_index','Team_A_points', 'Team_B_index', 'Team_B_points']
```

Construct random walk matrix M, team_index = 1,2,...,767

```python
M = np.zeros((769, 769))
```

```python
M.shape
```

```python
index = pd.DataFrame(np.arange(1,768))
name = pd.concat([index, name], axis=1, sort=False)
name.columns = ['name_id','team_name']
name.head()
```

```python
for i in range(0, data.shape[0]):
    row = data.iloc[i,:] #row is pandas series
    bottom_sum = row.Team_A_points+row.Team_B_points
    #team A wins
    if row.Team_A_points > row.Team_B_points:
        M[int(row.Team_A_index-1), int(row.Team_A_index-1)] += 1+row.Team_A_points/bottom_sum
        M[int(row.Team_B_index-1), int(row.Team_B_index-1)] += row.Team_B_points/bottom_sum
        M[int(row.Team_A_index-1), int(row.Team_B_index-1)] += row.Team_B_points/bottom_sum
        M[int(row.Team_B_index-1), int(row.Team_A_index-1)] += 1+row.Team_A_points/bottom_sum
    #team B wins
    elif row.Team_A_points < row.Team_B_points:
        M[int(row.Team_A_index-1), int(row.Team_A_index-1)] += row.Team_A_points/bottom_sum
        M[int(row.Team_B_index-1), int(row.Team_B_index-1)] += 1+row.Team_B_points/bottom_sum
        M[int(row.Team_A_index-1), int(row.Team_B_index-1)] += 1+row.Team_B_points/bottom_sum
        M[int(row.Team_B_index-1), int(row.Team_A_index-1)] += row.Team_A_points/bottom_sum
```

```python
M
```

Normalize Matrix

```python
M_norm = M/np.sum(M, axis=1).reshape(-1,1)
```

```python
M_norm
```

```python
w0 = np.ones(767)*np.array(1/767).reshape(1,-1)
```

```python
w0.shape
```

## a)  Use wt to rank the teams by sorting in decreasing value according to this vector.  List the top 25 team names (see accompanying file) and their corresponding values in wt for t= 10,100,1000,10000.

```python
def markov_chain_rank(w0, t):
    """
    Return top 25 ranking using Markov Chain.
    
    @w0: initial state vector
    @t: t time at Markov Chain
    """
    #count = 0
    acc = np.matmul(w0,M_norm) #w1
    for i in range(t-1):
        acc = np.matmul(acc, M_norm)
        #count +=1
    
    x = np.arange(1,768)
    d = {'index':x, 'correspond_w':acc.tolist()[0]}
    df = pd.DataFrame(d)
    df_temp = df.sort_values(by=['correspond_w'],ascending=False)
    result_df = pd.concat([df_temp, name], axis=1, join='inner').head(25)
    result_df = result_df.drop(columns=['index'])[['name_id','team_name','correspond_w']]
    return result_df
```

**t = 10**

```python
result_df = markov_chain_rank(w0, 10)
result_df
```

```python
result_df_100 = markov_chain_rank(w0, 100)
result_df_100
```

```python
result_df_1000 = markov_chain_rank(w0, 1000)
result_df_1000
```

**t = 10000**

```python
result_df_10000 = markov_chain_rank(w0, 10000)
result_df_10000
```

## b)

Use functions to dervie w_inf (stationary distribution)

```python
M_norm_T = M_norm.transpose()
```

```python
eigen_value, eigen_vector = eigs(M_norm_T,1)
```

```python
eigen_value
```

```python
eigen_vector = eigen_vector.flatten()
```

```python
w_inf = eigen_vector/np.sum(eigen_vector)
```

```python
w_inf.shape
```

Plot Graph

```python
diff_lst = []
wt_acc = np.matmul(w0,M_norm)#w1
diff_lst.append(np.sum(abs(wt_acc - w_inf))) #1st diff between wt and w0
for i in range(10000-1):
    wt_acc = np.matmul(wt_acc,M_norm) #wt
    diff_lst.append(np.sum(abs(wt_acc - w_inf)))   
```

```python
plt.figure(figsize=(8,6))
x_axis = np.arange(1,10001)
plt.plot(x_axis, diff_lst)
plt.xlabel('t')
plt.ylabel('||wt-w_inf|| (first norm)')
plt.title('||wt-w_inf|| (first norm) by t')
```

Construct Matrix X

```python
X = np.zeros([3012,8447])
```

```python
X.shape
```

```python
document_index = 0
acc_temp=[]
for i in range(0, nyt.shape[0]):
    #print(document_index)
    row = nyt.iloc[i,:].tolist()[0].split(',')
    for item in row:
        one_word = item.split(':')
        X[int(one_word[0])-1, document_index] = int(one_word[1])
        acc_temp.append(int(one_word[0]))
    document_index += 1
```

Initialize matrix W and H: each value in W and H can be initialized randomly to a positive number, e.g., from a Uniform(1,2) distribution.

```python
W = np.random.uniform(1,2,(3012,25))
H = np.random.uniform(1,2,(25,8447))
```

Implement Matrix Factorization using divergence objective

```python
purple = X/np.matmul(W,H)
```

```python
purple.shape
```

```python
W_trans = W.transpose()
```

```python
W_norm = W_trans/np.sum(W_trans, axis=1).reshape(-1,1) #normalize rows
```

```python
W_norm.shape
```

```python
part2 = np.matmul(W_norm, purple)
```

```python
part2.shape
```

```python
new_H = np.multiply(H, part2)
```

```python
new_H.shape
```

```python
def NMF(W, H, num_iteration):
    """
    Return W, H, divergence objective funtion after num_iterations.
    
    W: initialized W
    H: initialized H
    num_iteration: number of iterations
    """
    obj_acc = []
    for i in range(0,num_iteration):
        #update H
        purple = X/(np.matmul(W,H)+1e-16)
        W_trans = W.transpose()
        W_norm = W_trans/np.sum(W_trans, axis=1).reshape(-1,1) #normalize rows
        part2 = np.matmul(W_norm, purple)
        new_H = np.multiply(H, part2)
        H = new_H
        #update W
        #recalculate purple
        purple = X/(np.matmul(W,H)+1e-16)
        H_trans = H.transpose()
        H_norm = H_trans/np.sum(H_trans, axis=0) #normalize columns
        middle = np.matmul(purple, H_norm)
        new_W = np.multiply(W, middle)
        W = new_W

        #update objective
        WH = np.matmul(W,H)
        obj = np.sum(np.multiply(np.log(1/(WH+1e-16)),X) + WH)
        obj_acc.append(obj)
    
    return W, H, obj_acc
```

```python
init_W = np.random.uniform(1,2,(3012,25))
init_H = np.random.uniform(1,2,(25,8447))
W_NMF, H_NMF, obj_lst = NMF(init_W, init_H,100)
```

```python
plt.figure(figsize=(8,6))
x_axis = np.arange(0,100)
plt.plot(x_axis, obj_lst)
plt.title('Divergence Objective of Non-Negative Matrix Factorization after 100 iterations')
plt.xlabel('iterations')
plt.ylabel('divergence objective')
```
